## Defining a function and applying it to the df subset

```python
# Define a function
def price_label (row):
    
    if row['prices'] <= 5:
        return 'Low range product'
    elif (row['prices'] >5) and (row['prices'] <=15):
        return 'Mid range product'
    elif (row['prices'] >15):
        return 'High range product'
    else: return 'Not enough data'
```

```python
# Applying the function
df['price_range'] = df.apply(price_label, axis=1)
# This will create a new column titled price range
```

```python
# Checking the sum of the various categories created
df['price_range'].value_counts(dropna = False)
```

## Using the LOC method

```python
# Applying the LOC function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Checking the sum of the various categories created
df['price_range_loc'].value_counts(dropna = False)
```

## Creating a new column for the two busiest and two slowest days

```python
# Finding out the busiest and slowest days of the week
df_ords_prods_merged['order_day_of_week'].value_counts(dropna = False)
```

With this, we can establish that Saturday (0) and Sunday (1) are the busiest days of the week and Tuesday (3) and Wednesday (4) are the slowest days of the week.

Saturday and Sunday will be designated as Busiest days 
Wednesday and Thursday will be designated as Least busy days 
All other week days will be designated as Regularly busy day


```python
# Using the for-loop statement
end_result = []

for value in df_ords_prods_merged["order_day_of_week"]:
  if value == 0 or value == 1:
    end_result.append("Busiest days")
  elif value == 4 or value == 3:
    end_result.append("Least busy days")
  else:
    end_result.append("Regularly busy day")
```

```python
end_result
```

```python
# Merging it with the dataframe df_ords_prods_merged
df_ords_prods_merged['busiest_days'] = end_result
```

```python
# Checking column for accuracy
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

## Deriving a column for Busiest Period of the day

```python
# Finding out the busiest and slowest hours of the day
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

Since there are 24 (0-23) variables for hours of the day, I will divide them into 3 sections with 8 values each:
Most orders - 9, 10, 11, 12, 13, 14, 15, 16
Average orders - 7, 8, 17, 18, 19, 20, 21, 22
Fewest orders - 0, 1, 2, 3, 4, 5, 6, 23

```python
# Using the for-loop statement
hour_of_day = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in [9, 10, 11, 12, 13, 14, 15, 16]:
    hour_of_day.append("Most orders")
  elif value in [0, 1, 2, 3, 4, 5, 6, 23]:
    hour_of_day.append("Fewest orders")
  else:
    hour_of_day.append("Average orders")
```

```python
hour_of_day
```

```python
# Merging it with the dataframe df_ords_prods_merged
df_ords_prods_merged['busiest_period_of_day'] = hour_of_day
```

```python
# Checking the column's frequency and accuracy
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```

When compared with the sum of the order_hour_of_day column, the value has remained consistent (32,404,859)
