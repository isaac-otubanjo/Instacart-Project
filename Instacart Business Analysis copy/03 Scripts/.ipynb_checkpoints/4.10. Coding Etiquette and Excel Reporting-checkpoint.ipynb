### Considering Security Implications

Looking at the info for the dataframe, we can see that there are columns that contain PII (personally identifiable information). The columns include the First Name and Last Name column. It would be advisable to drop these columns as they are not relevant to our final Analysis.

```python
df_merged = df_merged.drop(columns = ['First Name', 'Last Name'])
```

```python
# Verifying results
df_merged.shape
```

### Comparing Customer Behavior in Different Regions

```python
# Defining Regions
Region=[]
for value in df_merged['State']:
    if value in ['Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut','New York','Pennsylvania','New Jersey']:
        Region.append('Northeast')
    elif value in ['Wisconsin','Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri']:
        Region.append('Midwest')
    elif value in ['Delaware','Maryland','District of Columbia','Virginia','West Virginia','North Carolina','South Carolina','Georgia','Florida','Kentucky','Tennessee','Mississippi','Alabama','Oklahoma','Texas','Arkansas','Louisiana']:
        Region.append('South')
    else:
        Region.append('West')
```

```python
# Create a new column containing the regions
df_merged['Region'] = Region
```

```python
df_merged['Region'].value_counts(dropna = False)
```

```python
# Crossing the variable with the spending flag
spend_habit = pd.crosstab(df_merged['Region'], df_merged['spending_flag'],dropna = False)
```

```python
# Plotting a bar chart with both variables
bar_spend_habit = spend_habit.plot.bar()
plt.title('Spending Habit Per Region')
```

The ratio of high to low spenders seem to be consistent across all regions indicating that Instacart has more low spenders than high spenders. Also there are more high and low spenders overall in the South than other regions.

```python
# Exporting bar chart
bar_spend_habit.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'Spending Habit Per Region.png'))
```

### Creating Profiling Variables based on Age, Income and Marital Status

```python
# For Age
df_merged['Age'].value_counts(dropna = False)
```

There are 64 unique values in the Age column (18-81). This would be divided into 3 categories: Young Adult, Middle Aged Adult and Old Adult.

```python
# Creating age flags for the 3 categories
df_merged.loc[df_merged['Age'] >= 40 & (df_merged['Age'] < 60), 'Age_category'] = 'Middle_aged_adult'
```

```python
df_merged.loc[df_merged['Age'] >= 60, 'Age_category'] = 'Old_adult'
```

```python
df_merged.loc[df_merged['Age'] < 40, 'Age_category'] = 'Young_adult'
```

```python
# Verifying results
df_merged['Age_category'].value_counts(dropna = False)
```

All customers have been grouped based on age.

```python
# For income, we can create a visualization to check the distribution
income_dis = df_merged['Income'].plot.hist(bins = 20)
plt.xlabel("Income")
plt.title("Income Distribution")
```

```python
# Exporting histogram
income_dis.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'Income Distribution.png'))
```

Using Statista.com, we can grade the incomes into 3 categories:
Low income (under 49,999), Median income (50,000 to 149,999) and High income (over 149,999).

```python
# Creating income flags for the 3 categories
df_merged.loc[df_merged['Income'] >= 50000 & (df_merged['Income'] < 149999), 'income_category'] = 'Median_income'
```

```python
df_merged.loc[df_merged['Income'] >= 149999, 'income_category'] = 'High_income'
```

```python
df_merged.loc[df_merged['Income'] <= 49999, 'income_category'] = 'Low-income'
```

```python
# Verifying results
df_merged['income_category'].value_counts(dropna = False)
```

All customers have been grouped based on income

```python
# For goods in the department column, we need to import the department dataframe
df_dept = pd.read_csv(os.path.join(path,'02 Data','Prepared Data','departments_wrangled.csv'),index_col = False)
```

```python
df_dept.head()
```

```python
# Renaming the "Unnamed: 0" column

df_dept.rename(columns={'Unnamed: 0':'department_id'}, inplace=True)
```

```python
# Merge with the df_merged dataframe on department_id
df_merged = df_merged.merge(df_dept, on = 'department_id')
```

```python
# Verifying result
df_merged.head()
```

```python
# Creating a bar chart to profile goods in the department column 
dept_bar_chart = df_merged['department'].value_counts().plot.bar()
plt.title('Order Numbers by Department')
```

```python
# Exporting the bar chart
dept_bar_chart.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'dept_bar_chart.png'))
```

```python
# To create a profile, we can aggregate some descriptive statistis for the department column 
crosstab_depts = df_merged.groupby('department').agg({'order_id':['count'],'days_since_last_order':['mean'],  'prices':['mean']})
```

```python
crosstab_depts
```

```python
# Creating customer profiles based on the highest and lowest average price
# Highest average price is Dairy Eggs and Lowest is Snacks
# Creating a flag to separate egg buyers from non buyers
df_merged.loc[df_merged['department_id'] == 16, 'customer'] = 'egg_buyer'
```

```python
df_merged.loc[df_merged['department_id'] != 16, 'customer'] = 'non_egg_buyer'
```

```python
# We can identify what customers are egg buyers using user id
egg_column = df_merged.loc[df_merged['customer'] == 'egg_buyer', 'user_id'].unique()
```

```python
df_merged.loc[df_merged['user_id'].isin(egg_column), 'customer'] = 'egg_buyer'
```

```python
df_merged['customer'].value_counts()
```

```python
# With the lowest average price which is snacks
df_merged.loc[df_merged['department_id'] == 19, 'customer_snack'] = 'snack_buyer'
```

```python
df_merged.loc[df_merged['department_id'] != 19, 'customer_snack'] = 'non_snack_buyer'
```

```python
# We can identify what customers are snack buyers using user id
snack_column = df_merged.loc[df_merged['customer_snack'] == 'snack_buyer', 'user_id'].unique()
```

```python
df_merged.loc[df_merged['user_id'].isin(snack_column), 'customer_snack'] = 'snack_buyer'
```

```python
df_merged['customer_snack'].value_counts()
```

All customers have been grouped based on department goods

```python
# For number of dependants, we can check for the various categories in the marital status column
df_merged['Marital Status'].value_counts()
```

```python
# Creating a family status flag
df_merged.loc[(df_merged['Marital Status'].isin(['divorced/widowed', 'living with parents and siblings', 'single'])) & (df_merged['No of dependants']==0), 'family_status_flag'] = 'Single with No Dependants'
```

```python
df_merged.loc[(df_merged['Marital Status']== 'married') & (df_merged['No of dependants']==1), 'family_status_flag'] = 'Married with No Dependants'
```

```python
df_merged.loc[(df_merged['Marital Status']== 'married') & (df_merged['No of dependants']>1), 'family_status_flag'] = 'Married with Dependants'
```

```python
df_merged.loc[(df_merged['Marital Status'].isin(['divorced/widowed', 'living with parents and siblings', 'single'])) & (df_merged['No of dependants']>=1), 'family_status_flag'] = 'Single with Dependants'
```

```python
# Verifying results
df_merged['family_status_flag'].value_counts()
```

All customers have been grouped based on marital status and number of dependants
