## Merging with the products dataframe

```python
df_prods
```

The most suitable way to combine df_merged_large with df_prods is using the product_id column which is present in both dataframes. 

Firstly, we need to drop the unnamed columns in both dataframes as they were produced as a result of exporting as csv

```python
# Dropping unnamed columns in df_prods
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
df_prods
```

```python
# Dropping unnamed and merge columns in df_merged_large
df_merged_large = df_merged_large.drop(columns = ['Unnamed: 0.1', '_merge'])
```

```python
df_merged_large
```

The dataframes can now be merged once the columns that were not needed have been removed

```python
# Merging both dataframes
df_ords_prods_merged = df_merged_large.merge(df_prods, on = 'product_id', indicator = True)
```

Confirming the results....

```python
df_ords_prods_merged
```

```python
# Confirming if this is a full merge
df_ords_prods_merged ['_merge'].value_counts()
```

This is a full merge
