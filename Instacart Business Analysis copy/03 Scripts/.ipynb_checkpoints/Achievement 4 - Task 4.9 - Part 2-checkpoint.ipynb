## Creating Visualisations

```python
# Creating Bar Charts
df_merged['order_day_of_week'].value_counts().plot.bar()
```

```python
# To sort by index position
df_merged['order_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# To add colours to the chart
df_merged['order_day_of_week'].value_counts().plot.bar(color =['purple', 'blue', 'red', 'yellow', 'orange', 'green', 'brown'])
```

```python
# Creating Histograms
df_merged['prices'].plot.hist(bins = 25)
```

```python
# Creating Scatterplots
sns.scatterplot(x = 'prices', y = 'prices',data = df_merged)
```

```python
# To check for outliers in the data due to the position of the points
df_merged.loc[df_merged['prices'] > 100]
# This will return prices higher than 100
```

```python
# Replacing all rows with prices higher than 100 with missing values
df_merged.loc[df_merged['prices'] > 100, 'prices'] = np.nan
```

```python
# Checking for the max value in the prices column
df_merged['prices'].max()
```

The outliers have now been taken care of. We can now create new visualisations. The new visualisations were created in the Exercise practise script which I have attached with the task.

### Creating Histogram for the order_hour_of_day column

```python
# Creating Histograms
hist_hour_of_day = df_merged['order_hour_of_day'].plot.hist(bins = 24)
```

The Histogram shows that the peak hours occur in the middle of the day (10:00 to 17:00). These represent the busiest hours for Instacart. The company could look to organise more adverts around the least busy hours (1:00 to 5:00).

```python
# Creating a bar chart for the loyalty flag column
bar_loyalty_flag = df_merged['loyalty_flag'].value_counts().plot.bar()
```

There is a majority of regular customers.

## Creating a sample dataframe for the line chart

```python
# Using a 70/30 split to create a list holding True/False values
np.random.seed(7)
dev = np.random.rand(len(df_merged)) <= 0.7
```

```python
dev
```

```python
# Store 70% of the sample in the dataframe big
big = df_merged[dev]
```

```python
# Store 30% of the sample in the dataframe small
small = df_merged[~dev]
```

```python
# Calculating the length of the dataframe
len(df_merged)
```

```python
# Confirming the results of the samples
len(big) + len(small)
```

```python
# Creating a new sample dataframe containing only the required columns
df_merged_2 = small[['order_hour_of_day', 'prices']]
```

### Creating a Line Chart

```python
# Creating line chart
line_rate = sns.lineplot(data = df_merged_2, x = 'order_hour_of_day',y = 'prices')
```

There doesn't seem to be any relationship between the expenditure and the hour of day as the busiest hours seem to have the lowest expenditure.

## Chart regarding marital status

```python
# Creating a bar chart for the loyalty flag column
bar_marital_status = df_merged['Marital Status'].value_counts().plot.bar()
```

Married people tend to spend more than every other category in this group.

### Checking for a connection between age and number of dependants

```python
# Creating a line chart between age and number of dependants
# Create a sample dataframe for the line chart
df_merged_3 = small[['No of dependants', 'Age']]
```

```python
# Creating line chart
line_rate_2 = sns.lineplot(data = df_merged_3, x = 'Age',y = 'No of dependants')
```

There is no relationship between Age and No of dependants.

### Checking if there is a connection between age and income

```python
# Creating a scatterplot
scatterplot_age_income = sns.scatterplot(x = 'Age', y = 'Income', data = df_merged)
```

The scatterplot shows that there is an increase in income with increase in age from around the 40 year mark. Overall, the income levels seem relatively stable from 20-40 and also from 40-80.
