## Merging df_customers with df_ords_prods

```python
# checking the columns in both dataframes to discover which columns are similar
df_ords_prods.head()
```

```python
df_customers.head()
```

```python
# Removing the missing values in the df_customers dataframe
# First create a subset
df_missing = df_customers[df_customers['First Name'].isnull() == True]
```

```python
# To confirm
df_missing
```

```python
# We can now drop the df_missing from the df_customers 
df_customers_cleaned = df_customers[df_customers['First Name'].isnull()== False]
```

```python
# To confirm
df_customers_cleaned
```

To confirm, there are now 194,950 rows which is a result of the df_customers (206,209) minus the df_missing (11,259).

```python
# To merge the newly cleaned customers dataframe to the ords_prods dataframe
df_ords_prods_merged = df_customers_cleaned.merge(df_ords_prods, on = 'user_id', indicator = True)
```

```python
df_ords_prods_merged.head()
```

```python
df_ords_prods_merged.shape
```

```python
# Checking to see if the merge was completed
df_ords_prods_merged['_merge'].value_counts()
```

The merge has been completed

```python
# Dropping the merge column from the merged dataframe
df_merged = df_ords_prods_merged.drop(columns = ['_merge'])
```
